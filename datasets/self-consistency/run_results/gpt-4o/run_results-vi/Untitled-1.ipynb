```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the graph
iterations = np.array([1, 2, 3, 4, 5])
f1_mini = np.array([0.657, 0.649, 0.662, 0.653, 0.644])
f1_nano = np.array([0.656, 0.655, 0.655, 0.658, 0.656])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(iterations, f1_mini, marker='o', label='gpt-4.1-mini (vi)')
plt.plot(iterations, f1_nano, marker='s', label='gpt-4.1-nano (en)')
plt.title('F1 Score Across Iterations')
plt.xlabel('Iterations')
plt.ylabel('F1 Score')
plt.ylim(0, 0.9)
plt.legend()
plt.grid(True)
plt.show()
```
